# DATASET 1: Speed Camera Violations

## Impoting Libraries

```python
from pymongo import MongoClient
import json
import pandas as pd
import datetime as dt
```

## Establishing connection with MongoDB

```python
client = MongoClient("mongodb://%s:%s@127.0.0.1" % ("dap", "dap"))
db = client['Traffic']
collection = db['Speed Camera Violations']
```

## Loading JSON Data

```python
with open("Speed_Camera_Violations.json", "r") as file:
    data = json.load(file)
    print("Successfully loaded")
```

## Storing data into MongoDB collection i.e., Speed

```python
def storejsondata():
    data_to_insert = data.get("results")
    for item in data_to_insert: 
        if isinstance(item, dict):         
            collection.insert_many([item])
        else:         
            print(f"Skipping item: {item}, not a valid dictionary.")
```

```python
storejsondata()
```

## Data Cleaning

### Converting data types of columns

```python
from datetime import datetime
```

```python
speed_violations['VIOLATION DATE'] = pd.to_datetime(speed_violations['VIOLATION DATE'])
speed_violations['X COORDINATE'] = pd.to_numeric(speed_violations['X COORDINATE'])
speed_violations['Y COORDINATE'] = pd.to_numeric(speed_violations['Y COORDINATE'])
speed_violations['LATITUDE'] = pd.to_numeric(speed_violations['LATITUDE'])
speed_violations['LONGITUDE'] = pd.to_numeric(speed_violations['LONGITUDE'])
```

```python
speed_violations.info()
```

### Removing the Location Column & _id column

```python
speed_violations.drop(columns='LOCATION',axis=1, inplace= True)
```

```python
speed_violations.drop(columns='_id',axis=1, inplace= True)
```

```python
speed_violations.head()
```

### Handling Missing Values

```python
def missing_imputation(x, stats = 'mean'):
    if (x.dtypes == 'float64') | (x.dtypes == 'int64'):
        x = x.fillna(x.mean()) if stats == 'mean' else x.fillna(x.median())
    return x
```

```python
speed_violations = speed_violations.apply(lambda x: missing_imputation(x))
```

```python
speed_violations.info()
```

### Handling Outliers

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
speed_violations.plot(kind = 'box')
```

```python
speed_violations['VIOLATIONS'] = speed_violations['VIOLATIONS'].clip(lower= speed_violations.VIOLATIONS.quantile(0.01),
                                                                    upper = speed_violations.VIOLATIONS.quantile(0.99))
speed_violations['X COORDINATE'] = speed_violations['X COORDINATE'].clip(lower= speed_violations['X COORDINATE'].quantile(0.01),
                                                                    upper = speed_violations['X COORDINATE'].quantile(0.99))
speed_violations['Y COORDINATE'] = speed_violations['Y COORDINATE'].clip(lower= speed_violations['Y COORDINATE'].quantile(0.01),
                                                                    upper = speed_violations['Y COORDINATE'].quantile(0.99))
speed_violations['LATITUDE'] = speed_violations['LATITUDE'].clip(lower= speed_violations['LATITUDE'].quantile(0.01),
                                                                    upper = speed_violations['LATITUDE'].quantile(0.99))
speed_violations['LONGITUDE'] = speed_violations['LONGITUDE'].clip(lower= speed_violations['LONGITUDE'].quantile(0.01),
                                                                    upper = speed_violations['LONGITUDE'].quantile(0.99))
```

# DATASET 3: Traffic Crashes Dataset

## Establishing connection with PostgreSQL

```python
import sqlalchemy as db
engine = db.create_engine('postgresql://postgres:root@localhost:5432/postgres')
```

```python
traffic_crashes = pd.read_csv('Traffic_Crashes.csv')
def write_dataset(name,dataset,engine):
    dataset.to_sql('%s' % (name),engine,index=True,if_exists='replace',chunksize=1000)

write_dataset('Crashes', traffic_crashes, engine)
print("Crashes tables created and data stored in that table into PostgreSQL")
```

```python
def read_dataset(name,engine):
    try:
        dataset = pd.read_sql_table(name,engine)
    except:
        dataset = pd.DataFrame([])
    return dataset

traffic_crash_sql = read_dataset('Crashes', engine)
```

#  Data Cleaning Process

### Handling missing values 

```python
missing_values = traffic_crash_sql.isnull().sum()
traffic_crash_sql['CRASH_DATE'] = pd.to_datetime(traffic_crash_sql['CRASH_DATE'], errors='coerce')

columns = ['POSTED_SPEED_LIMIT', 'INJURIES_TOTAL', 'CRASH_HOUR', 'CRASH_DAY_OF_WEEK', 'CRASH_MONTH']
outliers = traffic_crash_sql[columns].describe()

missing_values, outliers
```

```python
missing_values_percentage = traffic_crash_sql.isnull().mean() * 100

columns_to_drop = missing_values_percentage[missing_values_percentage > 50].index
traffic_crash_sql.drop(columns=columns_to_drop, inplace= True)

numerical_columns = traffic_crash_sql.select_dtypes(['float64', 'int64']).columns
traffic_crash_sql[numerical_columns] = traffic_crash_sql[numerical_columns].fillna(traffic_crash_sql[numerical_columns].mean())

categorical_columns = traffic_crash_sql.select_dtypes('object').columns
traffic_crash_sql[categorical_columns] = traffic_crash_sql[categorical_columns].fillna(traffic_crash_sql[categorical_columns].mode().iloc[0])

remaining_missing_values = traffic_crash_sql.isnull().sum()

missing_values_percentage, columns_to_drop, remaining_missing_values
```

## Crashes per Hours

```python
traffic_crash_sql['CRASH_DATE'] = pd.to_datetime(traffic_crash_sql['CRASH_DATE'])

# Subset DataFrame to include only crashes from 2018, 2019 and 2020, because other years contain only partial data.
traffic_crash_sql = traffic_crash_sql[(traffic_crash_sql['CRASH_DATE'] >= '2018-01-01') & (traffic_crash_sql['CRASH_DATE'] <= '2020-12-31')]
traffic_crash_sql = traffic_crash_sql[(traffic_crash_sql['CRASH_DATE'] >= '2018-01-01') & (traffic_crash_sql['CRASH_DATE'] <= '2020-12-31')]
```

```python
traffic_crash_sql['CRASH_Hour'] = traffic_crash_sql['CRASH_DATE'].dt.hour

# Plot number of crashes by hour of day
plt.figure(figsize=(10,5))
s = sns.barplot(data=traffic_crash_sql.groupby('CRASH_HOUR')['CRASH_RECORD_ID'].nunique().reset_index(), x='CRASH_HOUR', y='CRASH_RECORD_ID', palette='GnBu_r', linewidth=0)
s.set_title('Hourly Number of Reported Crashes in Chicago (2014 - 2023)', y=1.02, fontsize=14)
s.set_xlabel('Hour of Day', fontsize=13, labelpad=15)
s.set_ylabel('Number of Crashes', fontsize=13, labelpad=15)
plt.show()
```

## Daily crash count

```python
traffic_crash_sql['Date'] = traffic_crash_sql['CRASH_DATE'].dt.date

daily_crash_data = traffic_crash_sql.groupby('Date')['CRASH_RECORD_ID'].nunique().reset_index()

plt.figure(figsize=(15, 8))
s = sns.lineplot(data=daily_crash_data, x='Date', y='CRASH_RECORD_ID', color='blue', linewidth=2)
s.set_title('Daily Number of Reported Crashes', fontsize=16)
s.set_xlabel('Date', fontsize=14, labelpad=15)
s.set_ylabel('Number of Crashes', fontsize=14, labelpad=15)
s.set(yscale="log")
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_table = traffic_crash_sql.pivot_table(
    values='CRASH_RECORD_ID',
    index='CRASH_HOUR', 
    columns='CRASH_DAY_OF_WEEK', 
    aggfunc='count'
)

# Plot the heatmap
plt.figure(figsize=(25, 13))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Heatmap of Traffic Crashes by Crash hour and Day of Week')
plt.xlabel('Crash Hour')
plt.ylabel('Day Of Week')
plt.show()
```
